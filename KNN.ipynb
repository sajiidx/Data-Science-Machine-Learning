```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
col = ['sepal_length','sepal_width','petal_length','petal_width', 'class']
data = pd.read_csv('iris.data', header=None ,names = col)
data['class'] = data['class'].astype('category').cat.codes
```

```python
x = data.drop('class', axis=1)
y = data['class']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=3)
```

```python
def manhattan(a, b):
    return sum(abs(val1-val2) for val1, val2 in zip(a,b))
```

```python
for j in range(len(xtest)):
    new = []
    for i in range(len(xtrain)):
        xx = xtrain.iloc[i].tolist()
        yy = xtest.iloc[j].tolist()
        new.append(manhattan(xx, yy))
    xtrain['d' + str(j)] = new
```

```python
xtrain
```

```python
k = 3
```

```python
prediction = []
for i in range(len(xtest)):
    indexes = xtrain.sort_values('d' + str(i))['d1'][0:k].index
    predict = []
    for index in indexes:
        predict.append(ytrain[index])
    prediction.append(max(map(lambda val: val, set(predict))))
```

```python
print(prediction)
```

```python
print(ytest.tolist())
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(prediction, ytest.tolist())
```
